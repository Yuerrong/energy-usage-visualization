To fulfil the task of inputting and storing data, first of all, I initialize a dictionary to store data and let users input data and then store them into a dictionary. Secondly, I initialize lists to store data of dictionary in order to visualize them, and append every data into lists and then calculate the three total consumptions. Thirdly, to fulfil the task of visualization, I import matplotlib module and use data from lists to draw three graphs.

What I learned from AI are firstly, how to let graph y value starts from 0 (plt.ylim(bottom=0)), secondly, how to draw three lines in one line graph, thirdly, how to draw a pie chart, and let the proportion shows as x.y% form (autopct='%1.1f%%'). These are the three challenges I met and I use AI to learn from it and conquer it.

```python
# Initialize dictionary to store data
usage_data = {}


for i in range(12):
    month=input('Enter month:')
    elec_use=input('Enter the electricity usage:')
    gas_use=input('Enter the gas usage:')
    water_use=input('Enter the water usage:') # input data

    usage_data[month] = {
        'elec_use': elec_use,
        'gas_use': gas_use,
        'water_use': water_use
    } #store the data into the dictionary

#print(usage_data)

```

```python
print(usage_data) #check the data
```

```python
# Initialize lists to store data
months = []
elec_use = []
gas_use=[]
water_use=[]
for month, usage in usage_data.items(): # for loop to iterate all items in dictionary
    months.append(month)# append the key of dictionary to list month
    elec_use.append(float(usage['elec_use'])) # append the electricity usage value to list electricity usage
    gas_use.append(float(usage['gas_use']))    # append the gas usage value to list gas usage
    water_use.append(float(usage['water_use'])) # append the water usage value to list water usage

# Initialize a value to store data
elec_consump = 0
gas_consump = 0
water_consump = 0


for month, usage in usage_data.items():
    elec_consump +=float( usage['elec_use'])
    gas_consump += float( usage['gas_use'])
    water_consump += float( usage['water_use'])
 # calcute sums of three kinds of energy


#print(water_consump)
```

```python
import matplotlib.pyplot as plt  # import module


energy_types = ['Electricity', 'Gas', 'Water']
consumption=[elec_consump,gas_consump,water_consump]

# draw different types of energy's total consumption
plt.plot(energy_types, consumption)
plt.title('Comparison of Energy Types by Consumption')# write a title
plt.xlabel('Energy Type')
plt.ylabel('Total Consumption')
plt.ylim(bottom=0) # let y axis value start from 0 （cited in ChatGPT）


plt.show() #show the graph
```

```python
plt.plot(months, elec_use,  label='Electricity') # draw a line of monthly electricity usage
plt.plot(months, gas_use, label='Gas') # draw a line of monthly gas usage
plt.plot(months, water_use, label='Water') # draw a line of monthly water usage（cited in ChatGPT）
plt.title('Monthly Consumption Trends') #write the title
plt.xlabel('Months')
plt.ylabel('Different Usage')
plt.ylim(bottom=0)
plt.show() #show the graph
```

```python
sizes = [elec_consump, gas_consump, water_consump]# define the sizes of the pie chart
labels = ['Electricity', 'Gas', 'Water'] # define the labels of the pie chart


plt.pie(sizes, labels=labels, autopct='%1.1f%%') # draw a pie graph, and every proportion data shows as x.y% （cited in ChatGPT and update data by myself）
plt.title('Annual Total Consumption Breakdown by Energy Type') #write the title


plt.show() #show the graph
```
